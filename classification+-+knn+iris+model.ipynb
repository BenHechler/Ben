```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
X = iris.data 
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=4)
```

```python
iris_target_table=pd.DataFrame({'true_result':iris.target})
iris_data_table=pd.DataFrame({'sepal length':iris.data[:,0],'sepal width':iris.data[:,1],'petal length':iris.data[:,2],'petal width':iris.data[:,3]})
iris_dataset_table=iris_data_table.merge(iris_target_table,left_index=True,right_index=True)
iris_target_names_table=pd.DataFrame({'true_result_name':iris.target_names})
iris_full_target_table=iris_target_table.merge(iris_target_names_table,left_on='true_result',right_index=True,how='right')
iris_dataset_table=iris_data_table.merge(iris_full_target_table,left_index=True,right_index=True)
iris_dataset_table
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
knn.predict(X_test)
```

```python
t1=pd.DataFrame({'result_name':iris.target_names})
t2=pd.DataFrame({'result':knn.predict(X_test)})
t2=t2.merge(t1,left_on='result',right_index=True)
t3=pd.DataFrame({'sepal length':X_test[:,0],'sepal width':X_test[:,1],'petal length':X_test[:,2],'petal width':X_test[:,3]})
test_results=t2.merge(t3,left_index=True,right_index=True).sort_index()
test_results
```

```python
a=test_results.merge(iris_dataset_table,on=('petal length','petal width','sepal length','sepal width'),how='inner').drop_duplicates().reset_index(drop=True)
a['is true?']= a['result']==a['true_result']
a
```
